# Research Questions:
## Does temperature change alter riding behavior of CitiBike customers?
## Do these effects differ between men and women?

# Overview
This study examines the effects of outside temperature on Citibike ridership using data collected in July 2013 and January 2014.
Overall ridership is expected to be lower in January than July due to inclement weather, but do men and women respond to these changes differently?

# Hypotheses

### Null Hypothesis: No difference in the change in ridership levels between July and January for men and women.

## $H_{0}: \frac{\mu_{MJul}}{\mu_{MJan}} - \frac{\mu_{WJul}}{\mu_{WJan}} = 0$

### Alternative Hypothesis: A significant difference in the change in ridership levels between July and January for men and women.

## $H_{1}: \frac{\mu_{MJul}}{\mu_{MJan}} - \frac{\mu_{WJul}}{\mu_{WJan}} \neq 1$

Where $MJul$ and $MJan$ are the number of male riders in July and January, respectively, and $WJul$ and $WJan$ are the number of female riders in July and January, respectively.

$\alpha$ = 0.05

```python
jul = pd.read_csv('2013-07 - Citi Bike trip data.csv')
jan = pd.read_csv('2014-01 - Citi Bike trip data.csv')
```

```python
df = jul.append(jan)
df = df[df['gender'] > 0]
df['gender'] = df['gender'].replace({1 : 'Male', 2 : 'Female'})
df['starttime'] = pd.to_datetime(df['starttime'])
df['month'] = df['starttime'].dt.month
df['month'] = df['month'].replace({7 : 'July', 1 : 'January'})
df['day'] = df['starttime'].dt.date
df.head()
```

```python
tripCounts = df.groupby(['month', 'day', 'gender']).count()
tripCounts = pd.DataFrame(data =tripCounts['tripduration'])
tripCounts = tripCounts.rename(columns = {'tripduration' : 'trips'})
tripCounts = tripCounts.reset_index(level=['gender', 'month', 'day'])
tripCounts = tripCounts.sort_values(by=['day'])
tripCounts
```

```python
fig = plt.figure(figsize = (16,10))

g = sns.barplot(x = 'day', y = 'trips', hue = 'gender', palette = 'RdBu', alpha = 0.75, data = tripCounts)
g.set_xticklabels(labels = tripCounts['day'], rotation=45)
```

```python
tripW = tripCounts[tripCounts['gender'] == 'Female']
tripW = tripW.rename(columns = {'trips' : 'trips_female'})
tripW = tripW.drop('gender', axis=1)
tripM = tripCounts[tripCounts['gender'] == 'Male']
tripM = tripM.rename(columns = {'trips' : 'trips_male'})
tripM = tripM.drop('gender', axis=1)

genderCounts = tripW.merge(tripM)
genderCounts['trips_total'] = genderCounts['trips_female'] + genderCounts['trips_male']
genderCounts['pct_female'] = genderCounts['trips_female'] / genderCounts['trips_total']
genderCounts['pct_male'] = genderCounts['trips_male'] / genderCounts['trips_total']
genderCounts
```
