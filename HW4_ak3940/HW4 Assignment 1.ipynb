```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline

sns.set(style="darkgrid")
```

```python
def meanList(distName, distList):
    distMeans = {'distribution': [], 'mean': [], 'size': []}
    
    for i in range(100):
        sampMu = mean(distList[i])
        sampN = len(distList[i])
        distMeans['mean'].append(sampMu)
        distMeans['size'].append(sampN)
        distMeans['distribution'].append(distName)
    
    distMeans = pd.DataFrame.from_dict(distMeans)
    return distMeans
```

```python
mu = 500
sigma = 10
allMeans = pd.DataFrame({'distribution': [], 'mean': [], 'size': []})

#Gaussian Distribution
gaussList = []

for i in range(100):
    n = np.random.randint(10, 2000)
    samp = sigma * np.random.randn(n) + mu
    gaussList.append(samp)

gaussMeans = meanList('Gaussian', gaussList)
allMeans = allMeans.append(gaussMeans)


#Poisson Distribution
poissList = []

for i in range(100):
    n = np.random.randint(10, 2000)
    samp = np.random.poisson(lam = mu, size = n)
    poissList.append(samp)

poissMeans = meanList('Poisson', poissList)
allMeans = allMeans.append(poissMeans)

#Binomial Distribution
biList = []

for i in range(100):
    n = np.random.randint(10, 2000)
    samp = np.random.binomial(mu * 2, 0.5, size=n)
    biList.append(samp)

biMeans = meanList('Binomial', biList)
allMeans = allMeans.append(biMeans)

#Chi-Squared Distribution
chiList = []

for i in range(100):
    n = np.random.randint(10, 2000)
    samp = np.random.chisquare(mu, size = n)
    chiList.append(samp)

chiMeans = meanList('Chi-Squared', chiList)
allMeans = allMeans.append(chiMeans)
```

```python
#Log-Normal Distribution
logList = []

for i in range(100):
    n = np.random.randint(10, 2000)
    samp = np.random.lognormal(mu, size=n)
    logList.append(samp)
    
logMeans = meanList('Log-Normal', logList)
```

```python
p = sns.FacetGrid(allMeans, col = 'distribution', col_wrap = 2, hue = 'distribution')
p.map(sns.regplot, 'size', 'mean')
```

# Figure 1: Distribution Means & Sample Sizes

Larger sample sizes produce sample means that are closer to the population mean for all distributions.
This is in line with the central limit theorem and the idea that larger samples are more representative of their populations.

```python
sns.regplot(x = logMeans['size'], y = logMeans['mean'])
```

# Figure 2: Log-Normal Distribution Means & Sample Sizes

The log-normal distribution reflects similar characteristics to the non-log distributions, albeit on a log scale.

```python
#Plots
sns.distplot(allMeans['mean'])
```

# Figure 3: Combined Distribution of Sample Means

Combining the non-log distributions produces a Gaussian distribution that centers on the population mean $\mu = 500$.

```python
sns.distplot(logMeans['mean'])
```

# Figure 4: Log-Normal Distribution Mean

The log-normal distribution also has a normal distribution that centers near the log of the population mean  $\mu = 500$.
