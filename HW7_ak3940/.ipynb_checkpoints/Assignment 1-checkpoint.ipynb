```python
def censusAPIRequest(varName, geom, geomVals, largeGeom, largeGeomVals, myKey):
    root = 'https://api.census.gov/data/2016/acs/acs1'
    action = 'get={}'.format(varName) + ',NAME'
    geometry = 'for={}:{}&in={}:{}'.format(geom, geomVals, largeGeom, largeGeomVals)
    key = 'key={}'.format(myKey) #my API key
    url = root + '?' + action + '&' + geometry + '&' + key
    return url
```

```python
url = censusAPIRequest(keyNBB, 'public%20use%20microdata%20area', '*', 'state', '36', myKey)
resp = requests.request('GET', url).content
```

```python
pumaBB = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))
pumaBB.head()
```

```python
keyPop = 'B28002_001E'
url = censusAPIRequest(keyPop, 'public%20use%20microdata%20area', '*', 'state', '36', myKey)
resp = requests.request('GET', url).content
pumaPP = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))
pumaPP.head()
```

```python
bbpc = pd.read_csv( "~/PUIdata/ACS_16_1YR_GCT2801.ST50_with_ann.csv",
            usecols=["GCT_STUB.target-geo-id2","HC01","HC02"])
bbpc.drop([0], inplace=True)
bbpc = bbpc.rename(columns={'GCT_STUB.target-geo-id2': 'gid'})
bbpc['gid'] = bbpc['gid'].astype(int)
bbpc = bbpc[bbpc['gid'] > 99]
bbpc.gid[bbpc.gid < 200000] = bbpc.gid[bbpc.gid < 200000] - 100000
bbpc.gid[(bbpc.gid >= 200000) & (bbpc.gid < 300000)] = bbpc.gid[(bbpc.gid >= 200000) & (bbpc.gid < 300000)] - 198000
bbpc.gid[(bbpc.gid >= 300000) & (bbpc.gid < 400000)] = bbpc.gid[(bbpc.gid >= 300000) & (bbpc.gid < 400000)] - 297000
bbpc.gid[(bbpc.gid >= 400000) & (bbpc.gid < 500000)] = bbpc.gid[(bbpc.gid >= 400000) & (bbpc.gid < 500000)] - 396000
bbpc = bbpc[bbpc['gid'] < 5000]
bbpc.head()
```

```python
diff_ = pumaBB.merge(bbpc, right_on="gid", 
                     left_on="puma")[["pcBB", "HC01"]]


diff_["diff"] = np.abs(diff_["pcBB"] - diff_["HC01"].astype(float))
diff_.describe()
```

```python
pop = 'B00001_001E'
url = censusAPIRequest(pop, 'public%20use%20microdata%20area', '*', 'state', '36', myKey)
resp = requests.request('GET', url).content
pumaPop = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))
pumaPop.head()
```

# 5.2

```python
pumashp = pumashp.to_crs(epsg=2263)
linkNYC = linkNYC.to_crs(epsg=2263)
linkpp = gpd.sjoin(pumashp, linkNYC).groupby('puma')['link_site'].count()
linkpp = pd.DataFrame(linkpp)
linkpp = linkpp.reset_index(level='puma')
linkpp.head()
```

```python
pumashplc = pumashp.merge(linkpp, how='left', on='puma')
pumashplc['link_site'] = pumashplc['link_site'].fillna(0)
pumashplc.head()
```

```python
#NOTE: for this part I used kiosks per 1000 people instead of per 100 people, because the ratio was way too small.
pumashplc['linkNYCp1000p'] = pumashplc.link_site / pumashplc.B28002_001E  * 1000
pumashplc.describe()
```

```python
pumashplc.crs = from_epsg(4326)
pumashplcBase = pumashplc.plot(column = 'linkNYCp1000p', cmap = 'magma', figsize = (16,10), legend = True)
linkNYC.plot(ax = pumashplcBase, column = 'date_link_', cmap = 'OrRd', alpha = 0.5, markersize = 6)
```
