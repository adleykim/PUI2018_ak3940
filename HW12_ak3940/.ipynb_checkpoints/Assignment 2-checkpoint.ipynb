# Data Collection & Preparation

```python
import numpy as np
import os
import urllib
import geopandas as gpd
import pandas as pd
from sklearn.cluster import KMeans
import zipfile
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap
from sklearn.cluster import DBSCAN

%matplotlib inline
```

```python
flist = []

for i in range(4,10):
    f = 'zbp9' + str(i) + 'totals.zip'
    flist = np.append(flist, f)

for i in range(10):
    f = 'zbp0' + str(i) + 'totals.zip'
    flist = np.append(flist, f)
    
for i in range(5):
    f = 'zbp1' + str(i) + 'totals.zip'
    flist = np.append(flist, f)
```

```python
for i in range(len(flist)):
    url = 'https://github.com/fedhere/PUI2018_fb55/raw/master/HW12_fb55/' + flist[i]
    urllib.request.urlretrieve(url, flist[i])
    os.rename(flist[i], os.getenv('PUIDATA') + '/' + flist[i])
```

```python
nyczipsshp = gpd.read_file(os.getenv("PUIDATA") + "/ZIP_CODE_040114.shp")
nyczips = nyczipsshp.ZIPCODE.values
```

```python
df = pd.DataFrame({'zip': np.unique(nyczips)})

year = list(range(1994,2014))

for i in range(len(year)):
    zf = zipfile.ZipFile(os.getenv('PUIDATA') + '/' + flist[i])
    dfMerge = pd.read_csv(zf.open(flist[i].replace('.zip','.txt')), dtype = {0: str})
    dfMerge.columns = map(str.lower, dfMerge.columns)
    dfMerge = dfMerge[['zip', 'est']]
    dfMerge = dfMerge.rename(columns = {'est': '{}'.format(year[i])})
    df = df.merge(dfMerge, on = 'zip', how = 'left')
df = df.dropna()
df = df.set_index('zip', drop = True)
```

```python
df.head()
```

```python
for i in range(230):
    df.iloc[i] = (df.iloc[i] - df.iloc[i].mean()) / df.iloc[i].std()
df.head()
```

# K-means Clustering

```python
km = KMeans(n_clusters=5, random_state=444).fit(df) #5 clusters

km.labels_
zipClusters = df.reset_index()
zipClusters = pd.DataFrame(zipClusters['zip'])
zipClusters['cluster'] = km.labels_
```

```python
df = df.reset_index().merge(zipClusters, how = 'left')

df0 = df[df['cluster'] == 0].set_index('zip').drop(labels=['cluster', 'index'], axis = 1)
df1 = df[df['cluster'] == 1].set_index('zip').drop(labels=['cluster', 'index'], axis = 1)
df2 = df[df['cluster'] == 2].set_index('zip').drop(labels=['cluster', 'index'], axis = 1)
df3 = df[df['cluster'] == 3].set_index('zip').drop(labels=['cluster', 'index'], axis = 1)
df4 = df[df['cluster'] == 4].set_index('zip').drop(labels=['cluster', 'index'], axis = 1)

fig, ax = plt.subplots(3,2, figsize=(16,4))

df0.T.plot(ax = ax[0,0], color = 'k', alpha = .3, legend = False)
df0.mean().plot(ax = ax[0,0], color = 'y', linewidth = 2.5)
df1.T.plot(ax = ax[0,1], color = 'k', alpha = .3, legend = False)
df1.mean().plot(ax = ax[0,1], color = 'y', linewidth = 2.5)
df2.T.plot(ax = ax[1,0], color = 'k', alpha = .3, legend = False)
df2.mean().plot(ax = ax[1,0], color = 'y', linewidth = 2.5)
df3.T.plot(ax = ax[1,1], color = 'k', alpha = .3, legend = False)
df3.mean().plot(ax = ax[1,1], color = 'y', linewidth = 2.5)
df4.T.plot(ax = ax[2,0], color = 'k', alpha = .3, legend = False)
df4.mean().plot(ax = ax[2,0], color = 'y', linewidth = 2.5)
ax[0,0].axis('off')
ax[0,1].axis('off')
ax[1,0].axis('off')
ax[1,1].axis('off')
ax[2,0].axis('off')
ax[2,1].axis('off')
ax[0,0].set_title('Cluster 0')
ax[0,1].set_title('Cluster 1')
ax[1,0].set_title('Cluster 2')
ax[1,1].set_title('Cluster 3')
ax[2,0].set_title('Cluster 4')

plt.show()
```

## Figure 1: K-means Cluster Centers (Mean)

*Cluster means (yellow) plotted over time series data belonging to that cluster. Each time series represent the standardized number of establishments in a zip code for the years 1994-2013.*

```python
zipMap = nyczipsshp.merge(zipClusters, left_on = 'ZIPCODE', right_on = 'zip', how = 'left')

fig, ax = plt.subplots(figsize=(16,16), facecolor = 'gainsboro')
zipMap.plot(ax = ax, column = 'cluster', cmap = 'RdBu', edgecolor = 'gainsboro')
ax.axis('off')

legend_elements = [Patch(facecolor='darkred', edgecolor='darkred', label='Cluster 0'),
                   Patch(facecolor='salmon', edgecolor='salmon', label='Cluster 1'),
                   Patch(facecolor='white', edgecolor='gainsboro', label='Cluster 2'),
                   Patch(facecolor='skyblue', edgecolor='skyblue', label='Cluster 3'),
                   Patch(facecolor='midnightblue', edgecolor='midnightblue', label='Cluster 4')]

ax.legend(handles = legend_elements, loc=2)
plt.show()
```

## Figure 3: Map of K-means Clusters (k = 5)

*Zip codes clustered based on business trends between the years 1994-2013 using the k-means clustering algorithm. Each color represents a different cluster.*

# DBScan Clustering

```python
dfDB = df.set_index('zip').drop(label=['index', 'cluster'], axis = 1)

db = DBSCAN(eps=2, min_samples=2).fit(dfDB)
labels = (db.labels_).astype(int)
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
clusters = pd.Series([df[labels == i] for i in range(num_clusters)])
print('Number of clusters: %d' % num_clusters)

dfDB['cluster'] = labels
```

```python
dfDB0 = dfDB[dfDB['cluster'] == 0].drop(labels=['cluster'], axis = 1)
dfDB1 = dfDB[dfDB['cluster'] == 1].drop(labels=['cluster'], axis = 1)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,2))

dfDB0.T.plot(ax = ax1, color = 'k', alpha = .3, legend = False)
dfDB0.mean().plot(ax = ax1, color = 'y', linewidth = 2.5)
dfDB1.T.plot(ax = ax2, color = 'k', alpha = .3, legend = False)
dfDB1.mean().plot(ax = ax2, color = 'y', linewidth = 2.5)
ax1.axis('off')
ax2.axis('off')
ax1.set_title('Cluster 0')
ax2.set_title('Cluster 1')

plt.show()
```

# Figure 2: DBScan Cluster Centers (Mean)

*Time series of business activity in zip codes between 1994-2013 and clustered using the DBScan algorithm. The cluster center, indicated by the mean, is displayed in yellow.*

```python
zipClustersDB = dfDB.reset_index()
zipClustersDB = zipClustersDB[['zip', 'cluster']]

zipMapDB = nyczipsshp.merge(zipClustersDB, left_on = 'ZIPCODE', right_on = 'zip', how = 'left')

fig, ax = plt.subplots(figsize=(16,16), facecolor = 'gainsboro')
zipMapDB.plot(ax = ax, column = 'cluster', cmap = ListedColormap(['whitesmoke', 'darkred', 'midnightblue']), edgecolor = 'gainsboro')
ax.axis('off')

legend_elements = [Patch(facecolor='whitesmoke', edgecolor='gainsboro', label='Outliers'),
                   Patch(facecolor='darkred', edgecolor='darkred', label='Cluster 0'),
                   Patch(facecolor='midnightblue', edgecolor='midnightblue', label='Cluster 1')]

ax.legend(handles = legend_elements, loc=2)
plt.show()
```

## Figure 4: Map of DBScan Clusters
*Zip codes clustered based on business trends between the years 1994-2013 using the DBScan clustering algorithm.* 

# Comparison of K-means (k = 5) & DBScan

DBScan's two clusters produced a far clearer division than the five k-means clusters. This is because the k-means clusters are more arbitrarily selected. In this situation, where the number of clusters are unknown, DBScan appears to be the more appropriate model to use. Moreover, DBscan also includes outliers in its output, while k-means does not. This allows for a safer interpretation of the results, rather than forcing the creation of clusters where clusters do not exist.

# Interpretation

The five k-means clusters aggregate zip codes based on business trends as represented by the number of establishments in a given zip code for the years 1994-2013. Clusters 0 and 2 display a positive change in the number of establishments over time, cluster 1 displays a decrease in the number of establishments over time, and clusters 3 and 4 remain relatively stagnant.

DBScan's two clusters identify positive and negative timeseries trends among the zip codes. Within these two clusters there is little deviation from the mean. This is due to the lack of differentiation between business trends for individual zip codes are related to overall economic trends in NYC. For this reason we observe that the majority of zip codes experience a growth in the number of establishments alongside overall growth in NYC's economy. Certain zip codes that experience a downturn in the number of establishments are captured by the DBScan algorithm, but that doesn't necessarily mean the reasons for these downturns are the same. 

Both models have limited explanatory power and should not be used to assess overall economic health for the individual zip codes.
